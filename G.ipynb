```python
def Read_Two_Column_File(file_name):
    with open(file_name, 'r') as data:
        x = []
        y = []
        print(data)
        for line in data:
            p = line.split()
            x.append(float(p[0]))
            y.append(float(p[1]))

    return x, y

x1, y1 = Read_Two_Column_File('So data.txt')

x1 = x1[77:] # initial S0 data
y1 = y1[77:]

x2, y2 = Read_Two_Column_File('2_weights_weekend_06022023.txt')

x2 = x2[180:350] # S1 data
y2 = y2[180:350]
```

```python
import matplotlib.pyplot as plt
```

So data

```python
t = np.linspace(0, 80, 213)
def modeling(x0, a, phi, c):
  x = x0*np.exp(-a*t)*np.cos(((2*np.pi)/10)*t+phi)+c
  return x
m=modeling(100, 0.045, 0.03, 40)
plt.plot(t[50:], m[50:], 'g', label='modeling')
plt.plot(x1, y1, 'r', label='So')
plt.legend()
ma = average(m)
print(ma)
plt.axhline(40.32166302467734)
```

```python
m1=modeling(110, 0.035, 4.5, 54)
plt.plot(t[125:], m1[125:], 'g', label='modeling')
plt.plot(x2, y2, 'b', label='S1')
plt.legend()
ma1 = average(m1)
print(ma1)
plt.axhline(55.9)
```
